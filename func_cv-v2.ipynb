```python
from PIL import Image as image
from PIL import ImageFilter
from PIL import ImageEnhance
from IPython.display import display

img_path="sources/roger1.png"
img=image.open(img_path)
print(img.size)
img=img.convert("RGBA")
display(img)
```

```python
print(img.mode)
print(img.size)
```

```python
# 增強細節濾鏡
img=img.resize((500,500))
img_blurred = img.filter(ImageFilter.DETAIL) 
display(img_blurred)
img=img_blurred
```

```python
# 轉換成LA 加上透明度的模式
img_gray=img.convert("LA")
img=img_gray
display(img)
```

```python
# 對比畫圖
contour_img = img.filter(ImageFilter.CONTOUR).convert("RGBA")
# 增強輪廓圖片的對比度
enhancer = ImageEnhance.Contrast(contour_img)
contour_img = enhancer.enhance(13.0)
display(contour_img)
img=contour_img
```

```python
# 轉換成特定的大小
n=150

small_img=img.resize((n,n))
print(small_img.size)
display(small_img)
```

```python
# 計算像素切分點
bright_limit=0
for y in range(small_img.height):
    for x in range(small_img.width):
        bright_limit+=small_img.getpixel((x, y))[0]
bright_limit /= small_img.height*small_img.width
bright_limit*=0.8
print(bright_limit)
```

```python
# 圖案
pic="7"
# 製圖
ascii_art=""
for y in range(small_img.height):
    for x in range(small_img.width):
        brightness=small_img.getpixel((x, y))[0]
        if brightness > bright_limit:
            ascii_art += " "
        else:
            ascii_art += pic
    ascii_art+="\n"
```

```python
# 製圖-黑白顛倒
anti_ascii_art=""
for y in range(small_img.height):
    for x in range(small_img.width):
        brightness=small_img.getpixel((x, y))[0]
        if brightness*0.9 < bright_limit:
            anti_ascii_art += " "
        else:
            anti_ascii_art += pic
    anti_ascii_art+="\n"
```

```python
# 畫圖
print(ascii_art)
```
