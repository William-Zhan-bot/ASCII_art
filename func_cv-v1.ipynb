```python
from PIL import Image as image
from IPython.display import display

img_path="sources/doge.jpg"
img=image.open(img_path)
print(img.size)
display(img)
```

```python
print(img.mode)
print(img.format)
```

```python
# 
from PIL import ImageFilter
img_blurred = img.filter(ImageFilter.EDGE_ENHANCE) 
display(img_blurred)
img=img_blurred
```

```python
# LA 
img_gray=img.convert("LA")
display(img_gray)
```

```python
# 
n=100

small_img=img_gray.resize((n,n))
print(small_img.size)
display(small_img)
```

```python
# 
bright_limit=0
for y in range(small_img.height):
    for x in range(small_img.width):
        bright_limit+=small_img.getpixel((x, y))[0]
bright_limit /= small_img.height*small_img.width
bright_limit=bright_limit*0.8
print(bright_limit)
```

```python
# 
pic="7"
# 
ascii_art=""
for y in range(small_img.height):
    for x in range(small_img.width):
        brightness=small_img.getpixel((x, y))[0]
        if brightness > bright_limit:
            ascii_art += " "
        else:
            ascii_art += pic
    ascii_art+="\n"
print(ascii_art)
```
